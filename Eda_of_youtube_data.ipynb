Case Study 2: Cleaning & Organizing YouTube Earnings Data
Background: As a senior data engineer, you're responsible for cleaning and organizing the data so it’s ready for executive reporting. The management is particularly interested in sorting earnings, updating column names, and resetting index formats.

🧠 Questions:

Sort + Reset Index
Sort the data in descending order of highest_yearly_earnings, and reset the index after sorting.

Display the top 5 entries post-sort.

Rename + Replace
Rename the column highest_monthly_earnings to max_monthly_income and lowest_monthly_earnings to min_monthly_income.

Replace all occurrences of the Category Column - "People & Blogs" with "Personal Vlogs".

Set Index + Reset Index
Set Youtuber as the new index and display the data.

Then reset the index to bring it back to default.

```python
# Sort + Reset Index
# Sort the data in descending order of
# highest_yearly_earnings, and reset the index after sorting.

# Display the top 5 entries post-sort.
```

```python
df.sort_values(by='highest_yearly_earnings',ascending=False,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.head(5)
```

```python
# Rename + Replace
# Rename the column highest_monthly_earnings to max_monthly_income and lowest_monthly_earnings to min_monthly_income.

# Replace all occurrences of the Category Column - "People & Blogs" with "Personal Vlogs".
```

```python
df.rename(columns={'highest_monthly_earnings':'max_monthly_income','lowest_monthly_earnings':'min_monthly_income'},inplace=True)
```

```python
df.columns
```

```python
df['category'].head(10)
```
